```python
# Write a function that takes the radius of a circle and returns its area.

def area_circle(r):
    return 3.14 * r * r

r = float(input("Enter the radius of the circle: "))

print(f"The radius entered is: {r}")

print(area_circle(r))
```

```python
# 2. Write a function that accepts three numbers and returns the largest one.

def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

print(f"The numbers entered are: {a}, {b}, {c}")

print("The largest number is:", largest(a, b, c))
```

```python
# 3. Write a function that takes a string and returns its length
# without using built-in length functions.

def string_length(s):
    count = 0
    for i in s:
        count += 1
    return count

s = input("Enter a string: ")

print(f"The string entered is: '{s}'")

print(string_length(s))
```

```python
# 4. Write a function that checks whether a number is prime.

def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

n = int(input("Enter a number: "))

print(f"The number entered is: {n}")

print(is_prime(n))
```

```python
# 5. Write a recursive function to find the nth Fibonacci number.

def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n = int(input("Enter the position of the Fibonacci number to find: "))

print(f"The position entered is: {n}")

print(fibonacci(n))
```

```python
# 6. Write a recursive function to reverse a string.

def reverse_string(s):
    if s == "":
        return ""
    return reverse_string(s[1:]) + s[0]

s = input("Enter a string to reverse: ")

print(f"The string entered is: '{s}'")

print(reverse_string(s))
```

```python
# 7. Write a program to find the sum of all numbers from 1 to n using a for loop.

n = int(input("Enter a number: "))

print(f"The number entered is: {n}")

total = 0

for i in range(1, n + 1):
    total += i

print("Sum =", total)
```

```python
# 8. Write a program to count how many even numbers are present in a list.

numbers = [11,22,43,64,35,76,98]
count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("Even numbers =", count)
```

```python
# 9. Write a program to print all prime numbers between 1 and 100.

for num in range(2, 101):

    count = 0

    for i in range(1, num + 1):
        if num % i == 0:
            count += 1

    if count == 2:
        print(num)
```
